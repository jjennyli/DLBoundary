## Imports

## Structure

```python
def exp_model():
    inp = Input(shape=(1,3))
    
    # operators
    inp_nums = Lambda(lambda x: x[:,:,:2])(inp)
    
    # operation
    inp_op = Lambda(lambda x: x[:,:,2:3])(inp)
    
    # classifier layer
    cl_d1 = Dense(128)(inp_op)
    cl_d2 = Dense(2)(inp_op)
    
    # concat classifier output with numbers
    ca_d1_in = concatenate([inp_nums, cl_d1])
    ca2_d1_in = concatenate([inp_nums, cl_d2])
    
    # calculation layer
    ca_d1 = Dense(1024, activation='sigmoid')(ca_d1_in)
    ca_d2 = Dense(512)(ca_d1)
    ca_d3 = Dense(256)(ca_d2)
    ca_d4 = Dense(128)(ca_d3)
    ca_d5 = Dense(64)(ca_d4)
    
    # calculation layer2
    ca2_d1 = Dense(1024)(ca2_d1_in)
    ca2_d2 = Dense(512)(ca_d1)
    ca2_d3 = Dense(256)(ca_d2)
    ca2_d4 = Dense(128)(ca_d3)
    ca2_d5 = Dense(64)(ca_d4)
    
    # output layer
    o_in = concatenate([ca_d5, ca2_d5])
    o = Dense(1)(o_in)
    
    model = Model(inp, o)
    model.compile(
        loss="MeanSquaredError",
        metrics=['accuracy']
    )
    
    return model
```

```python
model = exp_model()
```

### Datasets

```python
num_examples = 20000
rstart = 1
rend = 10000

# MULT
mult_setX, mult_setY = ad.gen_data_mult(num_examples, rstart, rend)
mult_setX = mult_setX.reshape(num_examples, 1, 3)
```

```python
num_epochs = 1000
batch_size = num_examples

model.fit(
    mult_setX, mult_setY,
    batch_size=batch_size,
    epochs=num_epochs
)
```
